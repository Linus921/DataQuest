```python
# in autos column, strip out sring characters and convert to integer
# look at data and determine characters to strip out
print(autos["price"].head())
# following has to be replaced "$", "," in the data
autos["price"] = autos["price"].str.replace("$", "").str.replace(",", "").astype(int)
# check work below and confirm int64 as dtype
print(autos["price"].head())
```

```python
# in odometer column, strip out sring characters and convert to integer
# look at unique values
autos["odometer"].value_counts()
# comma and km need to be stripped out
print(autos["odometer"].head())
autos["odometer"] = autos["odometer"].str.replace(",", "").str.replace("km", "").astype(int)
# confirm change is to dtype int64
print(autos["odometer"].head())
# change column name to odometer_km
autos.rename({"odometer" : "odometer_km"}, axis = 1, inplace = True)
# confirm column name change and string to integer changes
autos.info()
```

```python
# investigating odometer_km and price columns in more detail
autos["odometer_km"].unique().shape
# 13 unique values exist for odometer_km
# look at min/max/median for the series
autos["odometer_km"].describe()
autos["odometer_km"].value_counts().sort_index(ascending=True)
# I do not see any real issues with this column's data
# I will now look at the price column
autos["price"].unique().shape
# there are 2357 unique prices values
autos["price"].describe()
autos["price"].value_counts().sort_index(ascending=True).head(10)
# there looks to be some unrealistic values in price column.  I should drop all rows where the price is greater than 150,000 or less than 500.
```

```python
# remove outliers in price less than 500 or greater than 150000
# create a boolean argument to remove those rows from dataframe
autos = autos[autos["price"].between(500,150000)]
autos["price"].value_counts().sort_index(ascending=True).tail()
```

___Data Justification___ In my analysis I chose to remove rows of data where the price appeared out of the norm for what I was expecting. There were a number of zero records and some close to one million. I decided to modify the data set and only keep those rows which were greater than 500 and less than 150000.

```python
# below line of code removes rows in data set which fall outside the 1900-2016 year range
autos = autos[autos["registration_year"].between(1900,2016)]
# running below command to make sure rows were removed correctly
# autos["registration_year"].value_counts().sort_index()
autos["registration_year"].value_counts(normalize=True, dropna=False).sort_index(ascending=False)
```

```python
# let me explore the brand column
autos["brand"].value_counts(normalize=True, dropna=False)
```

___Top Car Brands___ Looking at the above results, I plan to focus on teh top ten brands based on greatest percentile distribution.  Those brands are:  Volkswagen, BMW, Mercedes Benz, Ford, Opel, Audi, Fiat, Seat, Renault, and Puegot.  I now want to explore those brands in more detail by finding out the mean price for each one of them.  

```python
# create a variable using value counts to select the top ten brands
selected_brand = autos['brand'].value_counts(normalize=True).index[:10]
# pass the above variable into autos to select only those records from the data frame
mean_price_autos=autos.loc[autos['brand'].isin(selected_brand)]
# create variable that calculates the mean of the data frame
selected=mean_price_autos.groupby(['brand']).mean()
# execute selected to see the mean results
selected
```

___Mean Price Findings___ There is definitely a difference in price.  BMW, Mercedes Benz, and Audi are far away the more expensive cars in the data set.  Fiat, Opel, Puegot, and Renault are on the lower side of pricing.  Ford, Volkswagen, and Seat price is in the middle.

```python
# create empty dictionaries for mean price and mean mileage
mean_price={}
mean_mileage={}
# populate the dictionaries with values from the selected dataframe
mean_price=selected.iloc[:,0]
mean_mileage=selected.iloc[:,3]
#print(mean_price)
```

```python
# map mean price and mean mileage series into the autos dataframe
autos['mean_price']=autos['brand'].map(mean_price)
autos['mean_mileage']=autos['brand'].map(mean_mileage)
# review to see if it is as expected
autos.head(25)
```

___Data Findings after adding Mean Price & Mean Mileage___ the original theory holds regarding brand mean prices after adding mean mileage.  The mileage numbers are not that much different.  Had they been different that could have changed mean prices.  Thus, the data shows BMW, Audi, and Mercedez Benz are clearly the most expensive of the car brands.
