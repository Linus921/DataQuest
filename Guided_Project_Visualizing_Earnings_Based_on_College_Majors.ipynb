### Visualizing Earnings Based on College Majors ###



```python
import pandas as pd
import matplotlib.pyplot as plt
# import np.arange to evenly space bars
from numpy import arange
# enact Jupyter command so plots are displayed inline
%matplotlib inline
```

```python
# read in csv file, no encoding
recent_grads = pd.read_csv('recent-grads.csv')
```

***Explore the Data Set***

```python
print(recent_grads.iloc[0])
print('\n')
print(recent_grads.head())
print('\n')
print(recent_grads.tail())
print('\n')
recent_grads.describe()
```

```python
# drop rows with missing values so Matplotlib does not throw errors
recent_grads.info()
recent_grads = recent_grads.dropna()
recent_grads.info()
```

```python
# start exploring scatter plots for data correlations
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', 
                  title='Employed vs. Sample Size', figsize=(5,10))
```

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter', figsize=(5,10))
ax.set_title('Median vs. Sample Size')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', figsize=(5,10))
ax.set_title('Unemployment_Rate vs. Sample Size')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter', figsize=(5,10))
ax.set_title('Median vs. Full Time')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title=('Share Women vs. Unemployment Rate'), figsize=(5,10))
# ax.set_title('Share Women vs. Unemployment Rate')

```

```python
# ax = recent_grads.plot(x='Men', y='Median', kind='scatter', figsize=(5,10))
# ax.set_title('Median vs. Men')

recent_grads.plot(x='Men', y='Median', kind='scatter', figsize=(5,10)).set_title('Median vs. Men')  
# ax.set_title('Median vs. Men')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter', figsize=(5,10))
ax.set_title('Median vs. Women')
```

```python
# scatter matrix is another command that must be imported 
# it shows the relationship in histogram and scatter for 2-3 data points
# more than 3, becomes unreadable
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(6,6))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median','Unemployment_rate']], figsize=(10,10))
```

```python
recent_grads[:5].plot.bar(x='Major', y='Women')
```

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

```python
# create side by side bar of category major for men and women
recent_grads.groupby('Major_category')['Men', 'Women'].agg('sum').plot.barh()
```

```python
# create boxplot
recent_grads['Median'].plot.box(ylim=(0,80000))
plt.show()
recent_grads['Unemployment_rate'].plot.box()
```

```python
# another way to do the five histogram charts
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=40)
```

```python
# create hexbin chart for one of the dense scatterplot examples
ax = recent_grads.plot(x='Full_time', y='Median', kind='hexbin')
ax.set_title('Median vs. Full Time')
```
