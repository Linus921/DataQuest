## Guided Project: Analyzing CIA Factbook Data Using SQL

```python
# connect to database file
%capture
%load_ext sql
%sql sqlite:///factbook.db
```

```python
%%sql
select * from sqlite_master where type='table';
```

```python
%%sql
select * from facts limit 5;
```

Here are the descriptions for some of the columns:

* name - The name of the country.
* area - The total land and sea area of the country.
* population - The country's population.
* population_growth- The country's population growth as a percentage.
* birth_rate - The country's birth rate, or the number of births a year per 1,000 people.
* death_rate - The country's death rate, or the number of death a year per 1,000 people.
* area- The country's total area (both land and water).
* area_land - The country's land area in square kilometers.
* area_water - The country's waterarea in square kilometers.

```python
%%sql
SELECT 
MIN(population) min_pop,
MAX(population) max_pop, 
MIN(population_growth) min_growth, 
MAX(population_growth) max_growth 
FROM facts;
```

```python
%%sql
select name, population
from facts
where population = 0
```

```python
%%sql
select name, population
from facts
order by population DESC
limit 1;
```

```python
%%sql
SELECT * FROM facts
WHERE population == (
    SELECT MAX(population) FROM facts
);
```

```python
%%sql
SELECT * FROM facts
WHERE population == (
    SELECT MIN(population) FROM facts
);
```

```python
%%sql
select avg(population) avg_pop,
avg(area) avg_area
from facts;
```

```python
%%sql
select name, population
from facts
where population > 
(select avg(population) from facts)
order by population DESC;
```

```python
%%sql
select name, area
from facts
where area < (select avg(area) from facts);
```

```python
%%sql
select name
from facts
where name in 
(select name
from facts
where population > 
(select avg(population) from facts))
and name in (select name
from facts
where area < (select avg(area) from facts))
```
